```python
# imported dependencies
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
# compiled a list of all raw research data files (see RESEARCH for source details)
filenames = glob.glob("*with_ann.csv")
```

```python
# imported data from research CSV files into Pandas DataFrames, encoded into UTF-8, 
list_of_dfs = [pd.read_csv(filename, engine='c', sep=',', encoding = 'UTF-8', na_values=['*','**','***','N','-','',' ', '(X)'],low_memory=False, memory_map=True) for filename in filenames]
list_of_dfs_full = list_of_dfs
```

```python
# list_of_dfs
for dataframe, filename in zip(list_of_dfs, filenames):
    dataframe['filename'] = filename
    #save geolabels for later
    oldcols_df = list_of_dfs[0].filter(['GEO.id','GEO.id2','GEO.display-label'], axis=1)
    dataframe.drop(['GEO.id','GEO.id2','GEO.display-label','filename'], axis=1, inplace=True)
    
    dataframe.drop([0], axis=0, inplace=True)
#     filename.replace(regex=True,inplace=True,to_replace=r'_with_ann.csv',value=r'')
# list(my_dataframe)
```
